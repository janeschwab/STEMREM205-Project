```python
# Define a function to count overlaps between each sample
def overlap_counter(dataframe, column1, column2):
    """
    Args:
    dataframe: Pandas dataframe
    column1: Name of first column to compare (str)
    column2: Name of second column to compare (str)

    Returns:
    Given two Pandas dataframe columns with which we want to find the number of common elements,
    returns number of common elements between them, as well as the relative percentage
    of each column that is common.
    """

    from collections import Counter

    # Ensure given columns have no NaN values
    cleaned_col1 = [x for x in list(dataframe[column1]) if x == x]
    cleaned_col2 = [x for x in list(dataframe[column2]) if x == x]

    # Count occurrences of elements in each list
    counter1 = Counter(cleaned_col1)
    counter2 = Counter(cleaned_col2)

    # Calculate the intersection of the counters
    common_elements = counter1 & counter2

    # Sum of counts in the intersection
    res = sum(common_elements.values())

    # Calculate proportions of columns that are common
    proportion1 = res/len(counter1)
    proportion2 = res/len(counter2)

    # Print the result
    print(f"{column1} COMPARED TO {column2}")
    print("Number of common elements:", res)
    print(f"Percentage of {column1} that is common is {proportion1*100}%")
    print(f"Percentage of {column2} that is common is {proportion2*100}%")
    print("\n")
```
